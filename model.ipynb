```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB,CategoricalNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
gnb=GaussianNB()
bnb=BernoulliNB()
mlb=MultinomialNB()
csb=CategoricalNB()
```

```python
df=pd.read_excel("Raisin_Dataset.xlsx")
```

```python
df.head()
```

```python
x=df.drop(["Class"], axis=1)
y=df[["Class"]]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y, test_size=.2, random_state=42)
```

```python
model= {"Gaussian_NB":gnb,
        "BernoulliNB":bnb,
        "MultinomialNB":mlb
        }
```

```python
for name, models in model.items():
    print("model_name",model)
    model_data=models.fit(xtrain,ytrain)
    print(model_data.score(xtrain,ytrain))
    pred=models.predict(xtest)
    print(classification_report(pred, ytest))
    print(model_data.predict([[45264,290.58836,203.319142, 0.714455,48142, 0.656914,879.832]]))


 
```
